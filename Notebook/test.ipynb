```python
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/KeithGalli/Regression-Example/master/insurance.csv"
df = pd.read_csv(url)

df.head()
```

```python
display(df.info(),df.describe(),df.head(5))
```

```python
df.dropna(how = "all",inplace = True)
```

```python
#Make sure age is within the range 
#make sure childern is within the range 

df.loc[(df.age < 1) | (df.age > 100), "age"]=  np.nan
```

```python
df.loc[df.children < 0, "children"] = np.nan
```

```python
import re 

df["charges"] = df.charges.replace(r"[^0-9.]","",regex = True)
```

```python
df['sex'] = df.sex.replace(["female","woman"],"F")
df['sex'] =  df.sex.replace(["male","man"],"M")
df["region"] = df.region.str.lower()
df["age"] = df.age.fillna(df.age.median())
df["sex"] = df.sex.fillna(df.age.mode()[0])
df["bmi"] = df.bmi.fillna(df.bmi.mean())

```

```python
df = df.dropna(subset=["smoker"])
```

```python
df= df.dropna(how = "all")
```

```python

df["charges"] = pd.to_numeric(df["charges"], errors = "coerce")
```
