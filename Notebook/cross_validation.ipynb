```python
import numpy as np
from sklearn.datasets import make_classification

X, y = make_classification(
    n_samples=100,
    n_features=5,
    n_informative=3,
    n_redundant=0,
    n_clusters_per_class=1,
    weights=[0.9, 0.1],   # <-- IMBALANCE
    random_state=42
)

print("Class distribution:", np.bincount(y))
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score

kf = KFold(n_splits=5, shuffle=True, random_state=42)

for fold, (train_idx, val_idx) in enumerate(kf.split(X)):
    X_train, X_val = X[train_idx], X[val_idx]
    y_train, y_val = y[train_idx], y[val_idx]

    model.fit(X_train, y_train)
    preds = model.predict(X_val)

    print(
        f"Fold {fold} | "
        f"Val positives: {y_val.sum()} | "
        f"Accuracy: {accuracy_score(y_val, preds):.2f}"
    )
```

```python
from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for fold, (train_idx, val_idx) in enumerate(skf.split(X, y)):
    X_train, X_val = X[train_idx], X[val_idx]
    y_train, y_val = y[train_idx], y[val_idx]

    model.fit(X_train, y_train)
    preds = model.predict(X_val)

    print(
        f"Fold {fold} | "
        f"Val positives: {y_val.sum()} | "
        f"Accuracy: {accuracy_score(y_val, preds):.2f}"
    )
```

ðŸ” KFold vs StratifiedKFold â€” NOTES
ðŸ”¹ KFold
What it does

Splits data into K equal folds

Does NOT look at labels

Each fold is used once as validation

Uses

Regression

Classification only if classes are balanced

When label distribution does not matter

Properties

Random split (if shuffle=True)

Validation folds may have uneven class distribution

Code
from sklearn.model_selection import KFold

kf = KFold(n_splits=5, shuffle=True, random_state=42)

for train_idx, val_idx in kf.split(X):
    X_train, X_val = X[train_idx], X[val_idx]
    y_train, y_val = y[train_idx], y[val_idx]

Risk

âŒ For imbalanced classification, some folds may miss minority class

ðŸ”¹ StratifiedKFold
What it does

Splits data into K folds

Preserves class distribution in every fold

Uses y during splitting

Uses

Classification

Especially imbalanced datasets

Properties

Each fold has roughly same % of each class

More reliable validation metrics

Code
from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for train_idx, val_idx in skf.split(X, y):
    X_train, X_val = X[train_idx], X[val_idx]
    y_train, y_val = y[train_idx], y[val_idx]

Benefit

âœ… Model is evaluated on all classes in every fold
