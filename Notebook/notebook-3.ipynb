```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

df = pd.read_csv(url)


```

```python
display(df.info(),df.head(5), df.describe())
```

```python
df.Embarked.unique()
```

```python
df.drop(["PassengerId","Name","Ticket"],axis = 1 , inplace = True)
```

```python
#
```

```python
display(df.info(),df.describe())
```

```python
df.groupby("Survived")["Age"].median()
```

```python
df["Age"] = df["Age"].fillna(df.Age.median())
```

```python
df.groupby("Survived")["Embarked"].value_counts(normalize=True)
df["Embarked"] = df.Embarked.fillna(df.Embarked.mode()[0])
```

```python
df["Cabin"] = df.Cabin.fillna("Unknown")
```

```python
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
df.drop(["Cabin"],axis = 1,inplace = True)
```

```python
from sklearn.model_selection import train_test_split
X = df.drop("Survived",axis = 1)
y = df.iloc[:,[0]]

X_train,X_test,y_train,y_test = train_test_split(X, y ,train_size =0.8,random_state = 21)
```

```python
X_train,y_train
```

```python
from sklearn.preprocessing import StandardScaler,OneHotEncoder,OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier

scale_columns = ["Age","Fare"]
ohe_columns   = ["Sex", "Embarked"]
ordinal_columns = ["Pclass","SibSp", "Parch", "HasCabin"]

preprocessing = ColumnTransformer(transformers = [
    ("ordinal_encode", OrdinalEncoder(),ordinal_columns),
    ("one hot columns", OneHotEncoder( drop= "first", handle_unknown = "ignore"), ohe_columns),
    ("scale columns", StandardScaler(), scale_columns )
])

model_names = {
    "Logistic_Regression" : LogisticRegression(),
    "KNeighborsClassifier" : KNeighborsClassifier(),
    "Supportvectormachine" : SVC(kernel = "rbf"),
    "rf" : RandomForestClassifier(
            n_estimators=100
        ),
    "GradientBoostingClassifier" : GradientBoostingClassifier()
    }

for names,model in model_names.items():
    pipeline = Pipeline(
        steps = [
            ("preprocessing columns", preprocessing),
            ("model", model)
        ]
    )

    pipeline.fit(X_train,y_train)
    y_train_pred = pipeline.predict(X_train)
    y_test_pred = pipeline.predict(X_test)

    train_accuracy = accuracy_score(y_train,y_train_pred)
    test_accuracy = accuracy_score(y_test,y_test_pred)
    print(" model : {} , train_Accuracy : {}, test_accuracy : {}".format(names,train_accuracy,test_accuracy))
```
