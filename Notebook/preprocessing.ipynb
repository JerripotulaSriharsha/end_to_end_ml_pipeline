üëâ Left-skewed

Tree-based models don‚Äôt interpret numeric magnitude; they only use feature values to split data into subsets.

```python
from sklearn.preprocessing import LabelEncoder 

label_encoder_country = LabelEncoder()
label_encoder_purchased = LabelEncoder()
df["Country"] = label_encoder_country.fit_transform(df["Country"])
df["Purchased"] = label_encoder_purchased.fit_transform(df["Purchased"])
```

üå≥ TREE-BASED vs üìê LINEAR / DISTANCE MODELS (CATEGORICALS)
CORE DIFFERENCE (THIS IS EVERYTHING)
üå≥ Tree-based models

Think in IF / ELSE

Ask YES / NO questions

Do subsetting, not math

Do NOT care about numeric magnitude

Example split:

Is Country in {France, Germany} ?


‚û° Numbers are just IDs
‚û° LabelEncoder is OK

üìê Linear / Distance-based models

Do math

Multiply, add, subtract

Compute distance

Care about numeric magnitude

Example:

y = w * Country
distance = |2 - 0|


‚û° LabelEncoder creates fake meaning
‚û° One-Hot Encoding is REQUIRED

WHAT TO USE (MEMORISE)
Model type	Label Encoding	One-Hot Encoding
Tree-based	‚úÖ OK	‚úÖ OK
Linear / Logistic	‚ùå NO	‚úÖ YES
KNN / SVM	‚ùå NO	‚úÖ YES
ONE-LINE INTERVIEW ANSWER üéØ

‚ÄúTree-based models split data into subsets using rules, so label-encoded categories don‚Äôt introduce numerical bias, whereas linear and distance-based models require one-hot encoding to avoid artificial ordering.‚Äù

ONE-LINE MEMORY HOOK üß†
Trees ‚Üí rules ‚Üí subsets ‚Üí Label OK
Linear ‚Üí math ‚Üí distance ‚Üí One-Hot


That‚Äôs it.
If you can say this much clearly, you‚Äôre already ahead of most candidates.

What each method actually tells you (very clearly)
‚ùå Correlation

Tells: linear association between two numeric variables

Does NOT tell:

causation

non-linear effects

interaction effects

importance for classification

üëâ Useful for EDA sanity checks, not feature importance.

‚úÖ Logistic regression coefficients

Tell: direction and strength of linear effect on the label

Relationship type:

monotonic

additive

Good when:

model is linear

interpretability matters

üëâ Explains how a feature affects the label (linearly).

‚úÖ Tree feature importance

Tell: which features the model uses most to split

Captures:

non-linear relationships

interactions

Model-dependent:

importance ‚â† true causality

üëâ Explains what the model relies on.

‚úÖ Permutation importance (very important)

Tell: how much a feature affects prediction performance

Works for:

any model

any feature type

Measures:

impact on accuracy / AUC / loss

üëâ Best practical answer to
‚ÄúDoes this feature matter for predicting the label?‚Äù

‚úÖ SHAP

Tell:

global importance (overall)

local importance (per prediction)

direction + magnitude

Captures:

non-linearity

interactions

Closest thing to a full explanation.

üëâ Best tool to explain why the model predicts the label.

‚úÖ Mutual Information

Tell: dependency between feature and label

Does NOT assume:

linearity

Good for:

feature screening

early analysis

üëâ Answers: ‚ÄúIs there any signal at all?‚Äù

So‚Ä¶ are these enough?
‚úÖ YES ‚Äî together, they are sufficient to understand:

whether a feature is related to the label

whether the relationship is linear or non-linear

whether the feature actually helps prediction

how the model is using that feature

‚ùå NO single method is enough alone

You need at least one of:

model-based importance (tree / permutation / SHAP)

plus optional EDA (correlation / MI)

One sentence you should remember (and say in interviews)

‚ÄúFeature‚Äìlabel relationships in classification are best understood through model-based importance methods rather than correlation alone.‚Äù

That sentence is spot on.

Final simplified rule üß†
EDA        ‚Üí correlation, MI
Prediction ‚Üí permutation importance
Explanation‚Üí SHAP / coefficients
